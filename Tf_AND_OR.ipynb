```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense
import numpy as np
```

Creating dataset for an And Gate

```python
x = np.array([[0,0], [0,1], [1,0], [1,1]], "uint8") #input
y = np.array([[0], [0], [0], [1]], "uint8") #output
```

Creating a sequential model using tensorflow

```python
model = Sequential()
model.add(Dense(units = 1, activation = 'sigmoid', input_dim = 2)) #only one output
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics=['accuracy']) 
# model.summary()
```

```python
model.summary()
```

3 parameters to find (w0,w1,w2)


```python
model.fit(x, y, epochs = 750)  
```

```python
print(model.layers[0].get_weights())
```

```python
model.predict(x)
```

CREATING DATABASE FOR A XOR GATE

```python
x=[[0,0], [0,1], [1,0], [1,1]]
y = [[0], [1], [1], [0]]
```

```python
xor = Sequential()
xor.add(Dense(2, input_dim = 2, activation='sigmoid'))
xor.add(Dense(1, activation='sigmoid'))
xor.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
xor.summary()
```

```python
xor.fit(x, y , epochs = 1000)
```

```python
xor.get_weights()
```

```python
xor.predict(x)
```
